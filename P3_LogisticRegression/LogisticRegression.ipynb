```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("Samples/LR_Scores.csv")
X =   data.X.to_numpy()
Y =  (data.Y.to_numpy()>50).astype(np.float64)
print(X,Y)
```

```python
def CostF(x,w,b):
    return 1 / (1 + np.exp(- (np.dot(w, x) + b)))

def Gradient_Descent(w_old, b_old, points, l):
    m = len(points)
    grad_b = 0.0
    grad_w = np.zeros_like(w_old)
    for i in range(m):
        x = X[i]  
        y = Y[i]
        grad_b += ( CostF(x,w_old,b_old)-y) / m
        grad_w += x * (CostF(x,w_old,b_old)-y) / m 
    w = w_old - grad_w * l
    b = b_old - grad_b * l
    return w,b

w = np.zeros(1)
b = 0.0
for i  in range(500):
    w,b=Gradient_Descent(w,b,data,.3)
    if i%200==0:
        print(f"Loading {i}")

plt.scatter(X,Y,color="black")
plt.plot(list(range(0,11)),[CostF(x,w,b) for x in range(0,11)])
plt.show()
```

```python
predictions = CostF(4.9, w, b)

print(predictions > 0.5)
```
